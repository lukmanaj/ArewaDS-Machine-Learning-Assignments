# Dive into the beehive


Dig deeper into the honey dataset and build a notebook that can tell the story of the health of the bee population, state by state and year by year. 

Do you discover anything interesting about this dataset?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
honey = pd.read_csv('../data/honey.csv')
honey.head()
```

Let's look at how the stock varies over the years across the states.

```python
sns.relplot(data=honey,
            x='year',y='stocks',
            col='state',
            col_wrap=3,
            kind='line')
```

It's looking like in general the stocks fell. Though some state had higher stocks than others. Also, some states had peak years in between.

Now, looking at number of colonies and priceperlb over the years

```python
sns.relplot(data=honey,
            x='priceperlb',y='numcol',
            col='year',
            col_wrap=3,
            kind='line')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['numcol'], data=honey, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Price vs number of bee colonies over the years');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['priceperlb'], ax=ax2, color="r", 
                         label ='Price per lb', legend=False) 
sns.despine(right=False)
plt.ylabel('price per lb')
ax.figure.legend();
```

The price per lb continued to increase despite the fact the number of colonies seem to have stabilised around close to 60000. 

Finally, priceperlb and the stocks!

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['priceperlb'], data=honey, 
                        label = 'Price per lb', legend=False)
sns.despine()
plt.ylabel('# Prices')
plt.title('Price vs stock over the years');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['stocks'], ax=ax2, color="r", 
                         label ='stocks', legend=False) 
sns.despine(right=False)
plt.ylabel('stocks')
ax.figure.legend();
```

Looks like the stocks fell despite the continued increase in prices. Interesting. Something to look into further.
