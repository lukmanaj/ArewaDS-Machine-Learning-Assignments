```python
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('birds.csv')
birds.head()
```

```python
snow_goose_df = birds[birds['Name']=='Snow goose']
snow_goose_df.head()
```

```python
snow_goose_df.plot()
```

```python
minLength = snow_goose_df['MinLength']
maxLength = snow_goose_df['MaxLength']
category = snow_goose_df['Category']

plt.barh(category, maxLength)
plt.barh(category, minLength)

plt.show()
```

From the plot, maxlength of snow goose is around 80 (79) and the minlength is around 65 (64).

```python
conversation_status_count = snow_goose_df.value_counts(snow_goose_df['ConservationStatus'].values,sort=True)
plt.rcParams['figure.figsize'] = [6,12]
conversation_status_count.plot.bar()
```

Of course, they all have the same conservation status(Least Concern), after all, its the same bird!

```python
plt.title('Max Wingspan in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxWingspan'][i]
    plt.plot(x, y, 'bo')
    if birds['Name'][i] == 'Snow goose':
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=12)
    
plt.show()
```

Seeing how snow geese compare to the other birds in the dataset in terms of wingspan
